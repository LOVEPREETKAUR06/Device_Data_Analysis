# A1. Retrieve the model name, manufacturer, weight, price, and price currency for the device(s) based on the oem_id.

```python
import csv

def load_data_from_csv(file_path):
    """
    Load data from a CSV file into memory using the csv reader function.

    Args:
    - file_path (str): Path to the CSV file.

    Returns:
    - list: List of dictionaries representing each row of data.
    """
    data = []
    with open(file_path, 'r', newline='') as csvfile:
        csv_reader = csv.DictReader(csvfile)
        for row in csv_reader:
            data.append(row)
    return data

def retrieve_device_info_by_oem_id(data, oem_id):
    """
    Retrieve device information based on the oem_id.

    Args:
    - data (list): List of dictionaries representing each row of data.
    - oem_id (str): OEM ID to search for.

    Returns:
    - list: List of dictionaries containing device information matching the specified oem_id.
    """
    devices = []
    for device in data:
        if device['oem_id'] == oem_id:
            device_info = {
                'Model Name': device['model'],
                'Manufacturer': device['manufacturer'],
                'Weight': device['weight_gram'],
                'Price': device['price'],
                'Currency': device['price_currency']
            }
            devices.append(device_info)
    return devices

# Example usage:
csv_file_path = '/content/device_features.csv'
# oem_id_to_search = 'A135UZKDXAA'
oem_id_to_search = input("Enter OEM ID:")
data = load_data_from_csv(csv_file_path)
device_info = retrieve_device_info_by_oem_id(data, oem_id_to_search)
if device_info:
    print("Device(s) information for oem_id", oem_id_to_search, ":")
    for device in device_info:
        print(device)
else:
    print("No device found for oem_id", oem_id_to_search)
```

## b1. Identify the top 5 regions where a specific brand of devices was sold.


```python
def top_5_regions_for_brand(data, brand):
    """
    Identify the top 5 regions where a specific brand of devices was sold.

    Args:
    - data (DataFrame): DataFrame containing the device data.
    - brand (str): Brand name to search for.

    Returns:
    - list: Top 5 regions where the specified brand of devices was sold.
    """
    brand_data = data[data['brand'] == brand]  # Filter data for the specified brand
    region_counts = brand_data['market_regions'].value_counts().head(5)  # Count occurrences of each region
    return region_counts.index.tolist()  # Return top 5 regions

# Example usage:
csv_file_path = '/content/device_features.csv'
data = load_data_from_csv(csv_file_path)
brand_to_search = input("Enter Brand Name:")
top_regions = top_5_regions_for_brand(data, brand_to_search)

for region in top_regions:
    count = data[(data['brand'] == brand_to_search) & (data['market_regions'] == region)].shape[0]
    print(f"{region}: {count}")
```

## b2. Analyse the average price of devices within a specific brand, all in the same currency.

```python
def average_price_for_brand(data, brand):
    """
    Analyze the average price of devices within a specific brand, all in the same currency.

    Args:
    - data (DataFrame): DataFrame containing the device data.
    - brand (str): Brand name to analyze.

    Returns:
    - float: Average price of devices within the specified brand, all in the same currency.
    """
    brand_data = data[data['brand'] == brand].copy()  # Filter data for the specified brand and create a copy
    # Convert price column to numeric
    brand_data.loc[:, 'price'] = pd.to_numeric(brand_data['price'], errors='coerce')
    # Drop rows with missing prices
    brand_data = brand_data.dropna(subset=['price'])
    # Group by price currency and calculate average price
    average_price = brand_data.groupby('price_currency')['price'].mean().mean()
    return average_price


# Example usage:

brand_to_analyze = input("Enter Brand Name:")
average_price = average_price_for_brand(data, brand_to_analyze)
print("Average price of devices within the", brand_to_analyze, "brand, all in the same currency:", average_price)
```

```python
def analyze_device_weight_price_relationship(data):
    """
    Analyze the relationship between device weight and price.

    Args:
    - data (DataFrame): DataFrame containing the device data.

    Returns:
    - float: Correlation coefficient between device weight and price.
    """
    # Convert 'weight_gram' and 'price' columns to numeric
    data['weight_gram'] = pd.to_numeric(data['weight_gram'], errors='coerce')
    data['price'] = pd.to_numeric(data['price'], errors='coerce')

    # Drop rows with missing values in 'weight_gram' or 'price' columns
    data.dropna(subset=['weight_gram', 'price'], inplace=True)

    # Calculate correlation coefficient
    correlation_coefficient = data['weight_gram'].corr(data['price'])

    # Visualize the relationship between device weight and price
    sns.scatterplot(x='weight_gram', y='price', data=data)
    plt.xlabel('Device Weight (grams)')
    plt.ylabel('Price')
    plt.title('Relationship between Device Weight and Price')
    plt.show()

    return correlation_coefficient

# Example usage:

correlation_coefficient = analyze_device_weight_price_relationship(data)
print("Correlation coefficient between device weight and price:", correlation_coefficient)
```

## c3. Create separate charts illustrating the monthly average price trends (in GBP) for devices released in each year from 2020 to 2023. Each chart should focus on a specific year.

```python
df = data

# Convert 'released_date' to datetime
df['released_date'] = pd.to_datetime(df['released_date'], errors='coerce')

# Extract release year
df['release_year'] = df['released_date'].dt.year

# Filter data for years 2020 to 2023
df = df[df['release_year'].between(2020, 2023)]

# Group by year and month, calculate the monthly average price
monthly_avg_prices = df.groupby([df['release_year'], df['released_date'].dt.month])['price'].mean()

# Plotting
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

for year, ax in zip(range(2020, 2024), axes.flat):
    if year in monthly_avg_prices.index:
        monthly_avg_prices[year].plot(ax=ax, marker='o', linestyle='-')
        ax.set_title(f'Monthly Average Prices in {year}')
        ax.set_xlabel('Month')
        ax.set_ylabel('Average Price (GBP)')
        ax.grid(True)

plt.tight_layout()
plt.show()
```
